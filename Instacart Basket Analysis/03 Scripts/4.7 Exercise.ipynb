# Creating the “price_label” and “busiest_day” columns.

```python
df = ords_prods_merge[:1000000]
```

```python
df.shape
```

```python
#Define a function for price_label.
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
print(df.columns)
```

```python
# Function to apply to each column or row.
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# DataFrame with NA entries dropped from it or None if inplace=True.
df['price_range'].value_counts(dropna = False)
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
ords_prods_merge.rename(columns = {'order_dow' : 'orders_day_of_week'}, inplace = True)
```

```python
ords_prods_merge.drop(columns = ['eval_set'])
```

```python
# Which days do the most orders take place? 0=Saturday, count goes from there on.
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Creating a 'busiest day' column that shows busiest day, least busy, and regularly busy days.
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest days")
  elif value == 4:
    result.append("Slowest days")
  else:
    result.append("Normal days")
```

```python
ords_prods_merge['busiest_days'] = result
```

```python
result
```

```python
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
# Getting the value counts for each hour.
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
 # Dividing the hours of the day into 3 parts starting with the most orders.
hours = []
for value in ords_prods_merge["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        hours.append("Most orders")
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
        hours.append("Fewest orders")
    else:
        hours.append("Average orders")

ords_prods_merge['busiest_period_of_day'] = hours
```

```python
# Check frequency of busiest_period_of_day.
ords_prods_merge['busiest_period_of_days'].value_counts(dropna = False)
```
